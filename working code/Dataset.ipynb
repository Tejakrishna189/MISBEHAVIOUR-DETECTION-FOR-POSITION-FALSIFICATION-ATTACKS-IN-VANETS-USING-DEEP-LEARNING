```python
#importing required modules
import os
import json
import csv
#creating a list of column names
field_names = ['type','sendTime','sender','messageID','posX','posY','posZ','attackerType']
#creates a csv file named TestData to store all the message logs in json files
#writes column names
with open('TestData.csv','w') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames = field_names)
    writer.writeheader()
#creating a list to store all the folders(simulations)
liss=[]
#extracting all the folders(simulations)
for root, dirs, files in os.walk("C:\\Users\\kalyani\\Desktop\\Dataset"):
    liss=dirs
    break
#Iterate through each and every folder
for i in liss:
    #creating a list named ground_truth to store message logs of ground truth file
    ground_truth=[]
    print(i)
    #creating a list names jsonfiles to store message logs of json files
    jsonFiles=[]
    #select the folder
    folder_to_view ="C:\\Users\\kalyani\\Desktop\\Dataset\\"+i
    #extract the ground truth file
    ground_truth_file=os.listdir(folder_to_view)[1]
    #iterate through all messages in ground truth file
    for line in open(folder_to_view+"\\"+ground_truth_file):
        #load each line and append it to the list ground_truth
        a=json.loads(line)
        ground_truth.append(a)
    #iterate through each json file
    for filee in os.listdir(folder_to_view)[2:]:
        #iterate through each message log in json file
        for line in open(folder_to_view+"\\"+filee):
            #load each line
            a=json.loads(line)
            #if it is a GPS message, then skip
            if a['type']==2:
                continue
            #extract the position coordinates
            a['posX'] = float(a['pos'][0])
            a['posY'] = float(a['pos'][1])
            a['posZ'] = float(a['pos'][2])
            #remove unnecessary features
            del a['spd']
            del a['pos']
            del a['pos_noise']
            del a['spd_noise']
            del a['RSSI']
            del a['rcvTime']
            #add an extra feature named attacker type
            a['attackerType']=0
            #append the modified line into the list jsonfiles
            jsonFiles.append(a)
    #Iterate through the list ground_truth to get the attacker type
    for i in ground_truth:
        #continue if it is an attacker
        if i['attackerType']!=0:
            #find the equivalent message in jsonfiles and change the attacker type
            for j in jsonFiles:
                if i['messageID']==j['messageID']:
                    j['attackerType']=i['attackerType']
                    break
    #write all the message logs in jsonfiles into the TestData.csv file
    with open('TestData.csv','a+') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames = field_names)
        writer.writerows(jsonFiles)
```
